```python
import pickle as pkl
import numpy as np
import pandas as pd
import scipy.sparse as sp
```

```python
dataset = 'pubmed'
```

```python
names = ['graph', 'feature', 'labels']
objects = []
for i in range(len(names)):
    with open("{}.{}".format(dataset, names[i]), 'rb') as f:
        objects.append(pkl.load(f))
```

### Adjacency matrix
- sparse matrix type
- diagonal elements are removed

```python
adj = objects[0]; adj
```

- to numpy array

```python
adj_numpy = adj.todense(); adj_numpy
```

```python
print('edge num: {}'.format(np.count_nonzero(adj_numpy)/2))
```

- to dataframe

```python
adj_dataframe = pd.DataFrame(adj.todense()); adj_dataframe
```

### Feature matrix
- sparse matrix type

```python
feature = objects[1]; feature
```

- to numpy array

```python
feature_numpy = feature.todense(); feature_numpy
```

- to dataframe

```python
feature_dataframe  = pd.DataFrame(feature.todense()); feature_dataframe
```

### labels
- numpy array type

```python
labels = objects[2]; labels
```

```python
print('labels num: {}'.format(len(np.unique(labels))))
```

- to dataframe

```python
labels_dataframe = pd.DataFrame(labels); labels_dataframe
```

- Combine feature with labels

```python
feature_dataframe['labels'] = labels_dataframe; feature_dataframe
```
